# Plotting the similarity of Cities

Often, we hear that some city is totally like (or unlike) another city. Usually, such claims are followed by wild speculations on why this is the case. Most of the times, personal experience plays a major role ("I was mugged both in A and B", "The drinks here are spectacular", "My wife liked them a lot"). 
We want to add to this entertaining discussion by trying to visualize how similar cities are, based on the data Openstreetmap (OSM). OSM contains a lot of information on arts, shopping, entertainment and other things. 

The basic steps are the following:
- first, we retrieve the cities's geographical coordinates from OSM. Often this is the city center
- we then calculate a box (e.g. 1km^2) around this coordinate
- then, we retrieve all elements stored in OSM in this box, and filter out those that we are interested in
- afterwards, we do the real "similarity" stuff:
    - for each city, we put all the data to a high-dimensional vector. Each dimension represents a single category, e.g. "Shop-Shoes", "Amenity-Restaurant" or "Office-IT".
    - then, we use several dimensionality reduction techniques to reduce this high-dimensional vector to a low-dimensional vector. Specifically, we use a 2d-vector, so we can then nicely plot all the data in a diagram.
    
Cities which are close in the generated diagrams are then considered to be similar

# Getting the data

## Prepare stuff

OK, lets begin with the code. First we import all the stuff we need. "Overpass" is a small python wrapper libary, which provides the API to openstreetmap. See https://github.com/mvexel/overpass-api-python-wrapper for details.

```python
#import all required libraries
import overpass
import numpy
import math
import matplotlib.pyplot as plt
import numpy
import pylab
import cPickle as pickle
import os.path
import gzip
#initialize some stuff
#put plots right in this ipython notebook
%matplotlib inline 
api = overpass.API() #initialize overpass api
```

## Retrieve the cities

The first thing we have to define are the cities which we want to include in our little experiment. Attention: the script does not 
handle special characters so far :(

```python
cityNames=["Frankfurt am Main", "Stuttgart","Leipzig","Hamburg", 
           "Wien","London","New York","Paris","Madrid","Barcelona",
           "Wiesbaden","Mainz","Washington","San Francisco", "Athens","Sydney"]
```

Now, we start doing some real work. For each city name, we query OSM and retrieve all elements stored under this name. When we find an entry of type "place"="city", we use it. It will return the geographical coordinates (latitude and longitude) as well as some other information. 
As OSM queries can take some time, we store results using the pickle library (so that we dont have do do them again, if we, say, just add a city)

```python
cities={}
if os.path.isfile("cities.pklz"):
    f = gzip.open("cities.pklz",'r')
    cities=pickle.load(f)
    f.close()
notFound=[]
for cityName in cityNames:
    if cityName in cities.keys():
        print "Found city %s in saved cities"%(cityName) 
        continue
    response = api.Get('node["name"="'+cityName+'"]')
    cityFound=False
    i=0
    while not cityFound and i<len(response['elements']):
        if "place" in response['elements'][i]['tags']:
            if response['elements'][i]['tags']['place']=="city":
                cityFound=True;
                #save information
                try:
                    cities[cityName]={}
                    cities[cityName]['coords']=[float(response['elements'][i]['lat']),float(response['elements'][i]['lon'])]
                    if "is_in:iso_3166_2" in response['elements'][i]['tags'].keys():
                        cities[cityName]['isIn']=response['elements'][i]['tags']['is_in:iso_3166_2']
                    else:
                        cities[cityName]['isIn']="undefined"
                    cities[cityName]['population']=response['elements'][i]['tags']['population']
                    print "Found %s, %s in OSM"%(cityName,cities[cityName]['isIn'])
                except:
                    print "WARNING: Information on city %s did not include all required fields in OSM"%cityName
                    notFound.append(cityName)
                    continue
                
        i+=1
    if not cityFound:
        print "WARNING: City %s not found"%cityName
        notFound.append(cityName)
        continue

f = gzip.open('cities.pklz','wb')
pickle.dump(cities,f) 
f.close()

for city in notFound:
    cityNames.remove(city)
    
print "Done getting city coordinates"
```

Now, we need to transform the coordinates into an bounding box, which we will use to retrieve the POI data from OSM.
The coordinates we obtained are latitude and longitude, but we want to have a bounding box defined based on kilometers. So we have to transform it, using [this transformation](http://stackoverflow.com/questions/7477003/calculating-new-longtitude-latitude-from-old-n-meters).

```python
for city in cityNames:
    lat=cities[city]['coords'][0]
    lon=cities[city]['coords'][1]
    rEarth=6378 #km
    boxWidth=6 #km --> be careful, keep in mind that the area, and, hence, the number of nodes grows quadratically 
    deltaLat  =  (0.5*boxWidth / rEarth) * (180 / math.pi);
    deltaLon =  (0.5*boxWidth / rEarth) * (180 / math.pi) / math.cos(lat * math.pi/180);
    #from http://wiki.openstreetmap.org/wiki/Overpass_API:
    #"The order of values in the bounding box (51.249,7.148,51.251,7.152) is minimum latitude, minimum longitude, maximum latitude, maximum longitude (or South-West-North-East)."
    
    box=[lat-deltaLat,lon-deltaLon,lat+deltaLat,lon+deltaLon]
    cities[city]['boundingBox']=box 
print "Done calculating bounding boxes"
```

## Get the data

Now, we retrieve the data from OSM (via overpass). Again, we pickle the results, and use the stored results in case there are any (this step can take some time)

```python
tmpResponse={}
for city in cityNames:
    print city+":",
    pickleFileName=city+"-"+str(cities[city]['boundingBox'][0])+'-'+str(cities[city]['boundingBox'][1])+'-'+str(cities[city]['boundingBox'][2])+'-'+str(cities[city]['boundingBox'][3])+"-pickleDump.pklz"
    if os.path.isfile(pickleFileName): 
        print " (reading pickeled results)",
        f = gzip.open(pickleFileName,'r')
        response=pickle.load(f)
        f.close()
    else:
        response = api.Get('node('+str(cities[city]['boundingBox'][0])+','+str(cities[city]['boundingBox'][1])+','+str(cities[city]['boundingBox'][2])+','+str(cities[city]['boundingBox'][3])+')')
        f = gzip.open(pickleFileName,'wb')
        pickle.dump(response,f) 
        f.close()
    print " found %i nodes for %s"%(len(response['elements']),city)
    tmpResponse[city]=response
print "Done getting city data from OSM"
```

## Filter the data

OK, we have the data in place - which parts do we want to use? OSM uses "nodes", and we are interested not in all types of nodes, but only a subset (e.g., we dont want to have road intersections etc). See http://wiki.openstreetmap.org/wiki/Map_Features for a list of types.

So lets define two lists. `consideredTags` contains the top level node types a.k.a. tags that we will use. These types are fairly general. So the second list,  `blacklistedDetailTags` contains the second-level tags that we do *not* want to use. If we want to use tags of the type "natural", we add it to `consideredTags`. However, if we don't want to use the subtype "trees", so we add it to `blacklistedDetailTags` (trees actually seem to be something that has been mapped in some cities more than in others, so it might bias our experiment).

```python
consideredTags=["amenity","craft","historic","landuse",
                "leisure","military","office","public_transport","shop","sport",
                "tourism","waterway","natural"]
blacklistedDetailTags=["bicycle_parking","bicycle_rental","tree",
                       "stop_position","platform","entrance","yes"]
```

Alright, lets filter the data, and store only the information we want to use in the `cities` dictionary

```python
allTags={}
for city in cityNames:
    cities[city]['totalNodes']=0
    #check if it is entry we want to use
    for entryType in consideredTags:
        cities[city][entryType]={}
        allTags[entryType]={}
        for node in tmpResponse[city]['elements']:
            if "tags" in node.keys():
                if entryType in node['tags'].keys():
                    #only use non-blacklisted detail tags
                    if node['tags'][entryType] not in blacklistedDetailTags:
                        #count this node for this city
                        cities[city]['totalNodes']+=1
                        if node['tags'][entryType] in cities[city][entryType].keys():
                            cities[city][entryType][node['tags'][entryType]]+=1
                        else:
                            cities[city][entryType][node['tags'][entryType]]=1
                        #add type to total list
                        if node['tags'][entryType] in allTags[entryType].keys():
                            allTags[entryType][node['tags'][entryType]]+=1
                        else:
                            allTags[entryType][node['tags'][entryType]]=1
print "Done filtering data"
```

Furthermore, we add some (minor) filtering, by requiring a node type to occur at least in two cities. We then build the vector containing the information for each city. In this vector, we just add up the total numbers for each POI types for each city.

E.g., if we use the dimensions "hospital", "tree", and "bar", a vector
`[2 5 3]`
means that the considered city has two hospitals, four trees and three bars. To normalize the vectors, we divide each dimension by the total number of elements in the vector (i.e, each city vector will sum to one). Hence, our vector `[2 5 3]` will be normalized to `[0.2 0.5 0.3]`, as there is a total of ten POIs in the city (and 20% are hospitals).

```python
vectorLegend=[]
#reset vector for all cities
for city in cityNames:
    cities[city]['vector']=[]
print "Dimensions will be:"
for tag in allTags.keys():
    for tagDetail in allTags[tag].keys():
        #we want to have at least two of the node types (in total)
        if allTags[tag][tagDetail]>=2 and allTags[tag][tagDetail] not in blacklistedDetailTags:
            print " %s->%s"%(tag,tagDetail),
            vectorLegend.append([tag,tagDetail])
            for city in cityNames:
                increment=0
                if tagDetail in cities[city][tag].keys():
                    #increment=cities[city][tag][tagDetail]/float(cities[city]['totalNodes'])
                    increment=cities[city][tag][tagDetail]
                cities[city]['vector'].append(increment)
print ""
averageCity=numpy.zeros((len(cities[cityNames[0]]['vector']),))
for city in cityNames:
    cities[city]['vector']=numpy.array(cities[city]['vector'])/float(sum(cities[city]['vector']))
    averageCity+=cities[city]['vector']/float(len(cityNames))
print "Done building city vectors"
```

# Plotting the data

OK, now we can start plotting the data. First, we need to prepare the data, i.e, build a big matrix X with the vectors.

```python
X=[]
for city in cityNames:
    X.append(cities[city]['vector'])
X=numpy.array(X)
```

If we just add cities to the plots, they will be rather boring, because we have no clue why the algorithms placed two cities close to each other, or apart from each other. Hence, we add "dummy" cities, which each contain only a single type of POI, e.g., only trees or hospitals. We hope these prototypical cities tell us a bit more about the 2d space we are visualizing.

```python
R=numpy.zeros((len(vectorLegend),len(vectorLegend)))
#R=numpy.tile(averageCity,[len(vectorLegend),1])
for i in range(len(vectorLegend)):
    R[i,i]=1
```

This is the main plotting function, which we will use later on, once for each of the different dimensionality reduction techniques.

```python
def plotThis(X_r,R_r,plotName,plotMargin):
    #axis will go to plotMargin times largest/smallest value
    plt.figure()
    plt.title(plotName, fontsize=20)
    dataX=numpy.concatenate((X_r[:,0],R_r[:,0]))
    dataY=numpy.concatenate((X_r[:,1],R_r[:,1]))
    
    plt.scatter(dataX,dataY)
    minX=min(X_r[:,0])*plotMargin
    maxX=max(X_r[:,0])*plotMargin
    minY=min(X_r[:,1])*plotMargin
    maxY=max(X_r[:,1])*plotMargin
    outside=[]
    for i in range(len(vectorLegend)):
        if R_r[i,0]>minX and R_r[i,0]<maxX and R_r[i,1]>minY and R_r[i,1]<maxY:
            plt.annotate(vectorLegend[i][1],xy=(R_r[i,0],R_r[i,1]),color='red',fontsize=12)
        else:
            if abs(R_r[i,0])>abs(R_r[i,1]):
                mult=minX if R_r[i,0]<0 else maxX;
                xTmp=(R_r[i,0]/abs(R_r[i,0]))*abs(mult)
                yTmp=R_r[i,1]*(xTmp/R_r[i,0])
                rot=0
            else:
                mult=minY if R_r[i,1]<0 else maxY;
                yTmp=(R_r[i,1]/abs(R_r[i,1]))*abs(mult)
                xTmp=R_r[i,0]*(yTmp/R_r[i,1])
                rot=90
                
            outside.append(vectorLegend[i][1])
            plt.annotate(vectorLegend[i][1],xy=(xTmp,yTmp),color='blue',rotation=rot)
    
    for i in range(len(cityNames)):
        plt.annotate(cityNames[i],xy=(X_r[i,0],X_r[i,1]),fontsize=16)
    
    pylab.ylim([minY,maxY])
    pylab.xlim([minX,maxX])
    print "Outside of plot: ",
    print outside
```

## Principal Component analysis

Alright, lets transform the data using PCA. [PCA](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) projects data in a high-dimensional space to the hyperplane (here: in 2d) which explains most of the variance in the data. The 2d hyperplance is a linear combination of the input dimensions. In the plot below, we can, hence, infer that close-by points are similar, and points which are far away are dissimilar.

In the plot, the cities are labeled in black, "prototypes" are labeled in red, and prototypes with are outside of the plot area are labeld in blue at the edges (they might be far  out).

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
R_r=pca.transform(R)
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
plotThis(X_r,R_r,"PCA",1.1)
```

## t-SNE

t-Distributed Stochastic Neighbor Embedding (t-SNE) ([see here](http://lvdmaaten.github.io/tsne/), [or here](http://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html)) is a manifold learning algorithm. t-SNE focuses on pair-wise consistency, i.e,. it models local structures. The only "valid" interpretation in the picture below is therefore that adjacent dots are somehow similar.

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2, random_state=3)
allData=model.fit_transform(numpy.concatenate((X,R)))
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
X_r=allData[0:len(cityNames),:]
R_r=allData[len(cityNames):,:]
plotThis(X_r,R_r,"t-SNE",1.2)
```

# Some other techniques

## MDS

```python
from sklearn.manifold import MDS
model = MDS(n_components=2,metric=False)
allData=model.fit_transform(numpy.concatenate((X,R)))
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
X_r=allData[0:len(cityNames),:]
R_r=allData[len(cityNames):,:]
plotThis(X_r,R_r,"MDS",1.2)
```

## Locally linear embedding

```python
from sklearn.manifold import locally_linear_embedding
allData=numpy.concatenate((X,R))
myMethod='standard'
model = locally_linear_embedding(allData,n_neighbors=6,n_components=2,method=myMethod)
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
X_r=allData[0:len(cityNames),:]
R_r=allData[len(cityNames):,:]
plotThis(X_r,R_r,"Locally linear embedding ("+myMethod+")",1.2)
```

## Kernel PCA

```python
from sklearn.decomposition import KernelPCA
myKernel='sigmoid'
kpca = KernelPCA(n_components=2, kernel=myKernel)
X_r = kpca.fit(X).transform(X)
R_r=kpca.transform(R)
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
plotThis(X_r,R_r,"Kernel PCA ("+myKernel+")",1.1)
```

## Fast independent component analysis

```python
from sklearn.decomposition import FastICA
pca = FastICA(n_components=2)
X_r = pca.fit(X).transform(X)
R_r=pca.transform(R)
pylab.rcParams['figure.figsize'] = (15.0, 15.0)
plotThis(X_r,R_r,"FastICA",1.2)
```
